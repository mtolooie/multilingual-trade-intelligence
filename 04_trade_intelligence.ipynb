# 4. Trade Intelligence Results & Analysis

This notebook analyzes the final trade intelligence outputs derived from multilingual European institutional text.

We compare:
- Europarl (political discourse)
- JRC-Acquis (legal/regulatory discourse)

We evaluate:
- trade intensity scores
- topic-level trade signals
- cross-source differences

```python
import pandas as pd

df = pd.read_pickle("final_trade_intelligence.pkl")

print(df.shape)
print(df.columns)
df.head()
```

## Trade Intensity by Source

```python
df.groupby("source")["trade_intensity"].agg(["mean", "max", "min", "count"])
```

## Top Trade-Intensive Topics

```python
df.groupby("topic")["trade_intensity"].mean().sort_values(ascending=False).head(10)
```
