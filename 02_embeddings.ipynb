# 2. Multilingual Embedding Generation

This notebook transforms cleaned text into multilingual semantic vectors using a transformer model.

```python
import pandas as pd

df = pd.read_pickle("final_trade_intelligence.pkl")

print("Documents:", len(df))

print(df.columns)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "paraphrase-multilingual-MiniLM-L12-v2"
)

texts = df["clean_text"].tolist()

embeddings = model.encode(texts)

print("Embedding shape:", embeddings.shape)
```

```python
df["embedding"] = list(embeddings)

df.to_pickle("embedded_corpus.pkl")

print("Embeddings saved.")
```
